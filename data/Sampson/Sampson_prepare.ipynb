```python
import networkx as nx
import numpy as np
```

```python
G1 = nx.Graph(nx.readwrite.pajek.read_pajek('./Sampson0123_18.paj'))
nx.draw_networkx(G1)
print(G1.nodes())
adj1 = nx.adjacency_matrix(G1)
print(adj1.shape)
adj1.toarray()
```

```python
G = nx.Graph(nx.readwrite.pajek.read_pajek('./Sampson0123_25.paj'))
nx.draw_networkx(G)
G.nodes()
```

```python
adj2 = nx.adjacency_matrix(G)
adj2.toarray()
```

```python
index_list = np.zeros(18)
large_node_list = list(G.nodes())
for i, node in enumerate(G1.nodes()):
    index_list[i] = int(large_node_list.index(node))
print(index_list)
```

```python
index_list = list(np.array(index_list, dtype=int))
final_adj = adj2.toarray().copy()
final_adj[index_list][:,index_list] = final_adj[index_list][:,index_list] + adj1.toarray()
```
